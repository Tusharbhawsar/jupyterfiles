```python
import pandas as pd
import os
from threading import Thread
from tqdm import tqdm

video_dir = os.path.join('data', 'videos')
df = pd.read_csv('data/train_data/train_data.csv')

def crop_thread(video_path, start_time, duration, output_path):
    cmd = f'ffmpeg -v error -ss {float(start_time)} -i {video_path} -t {float(duration)} {output_path}'
    os.system(cmd)

threads = []
with tqdm(total=len(df), desc="Progress") as pbar:
    for idx, video_url, start_time, duration in zip(df.index, df.video_url, df.start_time, df.duration):
        output_name = video_url
        output_path = f"cropped_videos/{output_name}.mp4"
        video_path = os.path.join(video_dir, video_url)

        threads.append(Thread(target=crop_thread, args=(video_path, start_time, duration, output_path)))
        threads[-1].start()
        if len(threads) == 5:
            for i in threads:
                if i.is_alive():
                    i.join()
            threads = []
        pbar.update(1)
```

```python
import os
import shutil

# Set source and destination directories
dir_src = os.getcwd()
dir_dst = os.path.join(dir_src, 'out')

# Create the target directory if it doesn't exist
os.makedirs(dir_dst, exist_ok=True)

# Move all .png files from the source directory to the destination directory
for root, dirs, files in os.walk(top=dir_src):
    for file in files:
        if file.endswith(".png"):
            source_path = os.path.join(root, file)
            destination_path = os.path.join(dir_dst, file)

            # Handle the case where the destination file already exists
            count = 1
            while os.path.exists(destination_path):
                file_name, file_ext = os.path.splitext(file)
                new_filename = f"{file_name}_{count}{file_ext}"
                destination_path = os.path.join(dir_dst, new_filename)
                count += 1

            shutil.move(source_path, destination_path)

print("All .png files moved to 'out' directory.")
```
