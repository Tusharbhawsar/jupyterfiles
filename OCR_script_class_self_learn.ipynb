```python
import re
def get_pattren(input_str):
    pattern = r'\d{2}:\d{2}:\d+/\d{2}:\d{2}:\d+'

    # Find the first match in the input string
    match = re.search(pattern, input_str)

    # Process the match to truncate the seconds part
    if match:
        result = match.group()
        parts = result.split('/')

        for i, part in enumerate(parts):
            time_parts = part.split(':')
            if len(time_parts) == 3:
                # Truncate the seconds part to 2 digits
                time_parts[2] = time_parts[2][:2]
                parts[i] = ":".join(time_parts)

        truncated_result = "/".join(parts)
    else:
        truncated_result = input_str
#         print(truncated_result)
    return [truncated_result]
```

```python
import easyocr as eo
import glob
import natsort
import os
import re
import cv2
import json
import pandas as pd

class VideoTimeExtractor:
    def __init__(self, folder_path):
        self.folder_path = folder_path
        self.ocr = eo.Reader(lang_list=['en'])
        self.timing_json = {}

    def get_pattern(self, input_str):
        pattern = r'\d{2}:\d{2}:\d+/\d{2}:\d{2}:\d+'
        match = re.search(pattern, input_str)
        if match:
            result = match.group()
            parts = result.split('/')

            for i, part in enumerate(parts):
                time_parts = part.split(':')
                if len(time_parts) == 3:
                    time_parts[2] = time_parts[2][:2]
                    parts[i] = ":".join(time_parts)

            truncated_result = "/".join(parts)
        else:
            truncated_result = input_str

        return [truncated_result]

    def read_images(self):
        rev_data = natsort.natsorted(glob.glob(os.path.join(self.folder_path, "*.png")))
        return rev_data

    def extract_timestamps(self, img_path):
        filename = os.path.basename(img_path)
        img = cv2.imread(img_path, 0)
        img1 = img[1058:1076, 1650:1716]
        text = self.ocr.readtext(image=img1, mag_ratio=2, detail=0, allowlist=['1', '2', '3', '4', '5', '6', '7', '8', '9', '0', ':', '/', ])
        text = self.get_pattern(text[0])
        try:
            self.timing_json[filename] = text[0]
        except:
            self.timing_json[filename] = 'missed fetching the timing'

    def process_images(self):
        images = self.read_images()
        for image_path in images:
            self.extract_timestamps(image_path)

    def save_json(self, json_path):
        with open(json_path, "w") as outfile:
            json.dump(self.timing_json, outfile, indent=1)

    def format_time(self, time_str):
        formatted_time = "".join([time_str[i:i + 2] for i in range(0, len(time_str), 2)])
        return formatted_time

    def load_json(self, json_file):
        with open(json_file, "r") as fi:
            return json.load(fi)

    def create_output(self, output_file):
        data = self.load_json(output_file)
        output = []
        keys = list(data.keys())
        for i in range(0, len(keys), 2):
            img1 = keys[i]
            img2 = keys[i + 1]
            start_time = self.format_time(data[img1])
            end_time = self.format_time(data[img2])
            output.append(f"{start_time}-{end_time}")

        with open("final.json", "w") as tu:
            json.dump(output, tu, indent=1)
        return output

# Example usage:
folder_path = '/home/link-lap-24/Downloads/base/'
output_json_path = 'ann.json'

video_time_extractor = VideoTimeExtractor(folder_path)
video_time_extractor.process_images()
video_time_extractor.save_json(output_json_path)
output_data = video_time_extractor.create_output(output_json_path)
print(output_data)
```

```python
import os
import subprocess

class VideoProcessor:
    def __init__(self):
        self.output_folder = "output_frames"

        # Create output folder if it doesn't exist
        if not os.path.exists(self.output_folder):
            os.makedirs(self.output_folder)

    def trim_video(self, input_path, output_path, start_time, end_time):
        # Trim the video using ffmpeg
        cmd = [
            "ffmpeg",
            "-ss", str(start_time),
            "-i", input_path,
            "-t", str(end_time - start_time),
            "-c:v", "libx264",
            "-c:a", "aac",
            "-strict", "experimental",
            output_path
        ]
        subprocess.run(cmd)

    def extract_frames(self, input_video, output_folder):
        # Extract frames from the video using ffmpeg
        cmd = [
            "ffmpeg",
            "-i", input_video,
            "-vf", "fps=1",  # Extract 1 frame per second
            os.path.join(output_folder, "frame_%04d.jpg")
        ]
        subprocess.run(cmd)

# Example usage:
video_processor = VideoProcessor()

# Trim the video (assuming input.mp4 exists)
video_processor.trim_video("input.mp4", "trimmed_output.mp4", 10, 20)

# Extract frames from the trimmed video
video_processor.extract_frames("trimmed_output.mp4", video_processor.output_folder)
```

```python
from datetime import datetime

def check_timestamps(timestamps):
    for timestamp in timestamps:
        try:
            start_time, end_time = timestamp.split('-')
            start_datetime = datetime.strptime(start_time, '%H:%M:%S')
            print(start_datetime)
            end_datetime = datetime.strptime(end_time, '%H:%M:%S')

            if start_datetime >= end_datetime:
                print(f"Issue found: Start time {start_time} is greater than or equal to end time {end_time}")
            elif start_time > end_time:
                print(f"Issue found: Timestamp {timestamp} is in the wrong order")
        except ValueError as e:
            print(f"Issue found: {e} in timestamp {timestamp}")

# Given timestamps
timestamps = [
    "01:18:13-01:18:22",
    "01:19:57-01:19:59",
    "01:20:05-01:20:11",
    "01:20:15-01:20:16",
    "01:20:33-01:20:37",
    "01:20:58-01:21:00",
    "01:21:27-01:21:31",
    "01:21:40-01:21:45"
]

# Check timestamps for issues
check_timestamps(timestamps)
```
