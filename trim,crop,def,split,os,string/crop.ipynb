```python
#crop video

import os 
import natsort
import time

out_dir = "output/"
def crop_video(dir):
    stime = time.time()
    
    videos =  natsort.natsorted(os.listdir(dir))
    print(videos)
    
    if not os.path.exists(out_dir):
        os.mkdir(out_dir)
    
    for video in videos:
        cmd = f"""ffmpeg -i {dir + video} -ss 00 -to 5 -y {out_dir + video}"""
        print(cmd)
        os.system(cmd)
    
    total_time = time.time() - stime
    print(total_time, "total_time")
    
crop_video("input/")
    
```
