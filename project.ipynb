```python
#crop video without time

import os 
import natsort

out_dir = "output/" # /it will take inside the files

def crop_video(dir):
    
    videos =  natsort.natsorted(os.listdir(dir)) # sorting the files or for directly apply loop in folder
    print(videos)
    
    if not os.path.exists(out_dir):# if path not exit it will make path
        os.mkdir(out_dir)
    
    for video in videos:
        cmd = f"""ffmpeg -i {dir + video} -ss 00 -to 5 -c copy -y {out_dir + video}"""
        print(cmd)
        os.system(cmd)
        
crop_video("input/")
    
```

```python
frame_path='output/'

def trim_video(video_path):
    
    sorting=natsort.natsorted(os.listdir(video_path))
    print(sorting)
    
    if not os.path.exists(frame_path):
        os.mkdir(frame_path)
    
    for sort in sorting:
        cmd=f"""ffmpeg -i {video_path + sort} -ss 00 to 5 -c copy -y {frame_path + sort}"""
        print(cmd)
        os.system(cmd)
        
trim_video('input_data/')    
```
