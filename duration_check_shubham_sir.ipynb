```python
import pandas as pd

data = pd.read_csv('Cricket_Events_Clips-25-04-2023-10-37-00.csv')
```

```python
catch, throw = [], []

for n, i in enumerate(data['Tags']):
    if i == 'throw':
        tmp = {}
        tmp['clip_id'] = data['Source Url'][n].split('/')[-1].split('.')[0]
        tmp['url'] = data['Source Url'][n]
        tmp['startDuration'] = data['startDuration'][n]
        tmp['endDuration'] = data['endDuration'][n]
        throw.append(tmp)
    elif i == 'caught':
        tmp = {}
        tmp['clip_id'] = data['Source Url'][n].split('/')[-1].split('.')[0]
        tmp['url'] = data['Source Url'][n]
        tmp['startDuration'] = data['startDuration'][n]
        tmp['endDuration'] = data['endDuration'][n]
        catch.append(tmp)
```

# Catch

```python
td = []

for i in catch:
    print(i['url'])
    print(float(i['endDuration']),float(i['startDuration']))
    print(float(i['endDuration'])-float(i['startDuration']))
    td.append(float(i['endDuration'])-float(i['startDuration']))
```

```python
print("############################################")
print(f"min duration of catch - {min(td)}")
print(f"index number - {td.index(min(td))}")

clip = catch[td.index(min(td))]
for i in clip:
    print(f"{i} : {clip[i]}")
print("############################################")
print(f"max duration of catch - {max(td)}")
print(f"index number - {td.index(max(td))}")

clip = catch[td.index(max(td))]
for i in clip:
    print(f"{i} : {clip[i]}")
print("############################################")
```
