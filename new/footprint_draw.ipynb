```python
# for footprint

import cv2
import json
import os
from collections import deque

json_path = '/home/link-lap-24/track_net/examples/videos/ball_detections.json'

input_folder = '/home/link-lap-24/track_net/examples/videos/fr/'
output_folder = '/home/link-lap-24/track_net/examples/videos/out2'
os.makedirs(output_folder, exist_ok=True)

with open(json_path, 'r') as f:
    data = json.load(f)

# Track karne ke liye last 5-6 footprints store karne ke liye 
trail_points = deque(maxlen=6)  # Yeh last 6 points tak store karega

for frame_data in data:
    frame_number = frame_data['frame']
    x, y = frame_data['xy']
    
    image_path = os.path.join(input_folder, f"{frame_number}.jpg")
    image = cv2.imread(image_path)
    
    if image is None:
        print(f"Image load nahi ho rahi: {image_path}")
        continue
    
    # Current point ko trail_points me add karein
    trail_points.append((int(x), int(y)))
    
    # Har trail point par circle draw karein
    for i, point in enumerate(trail_points):
        # Color intensity decrease karne ke liye color change hota hai
        color = (0, 255 - i*40, 0)  # Green shade mein gradation
        cv2.circle(image, point, radius=5, color=color, thickness=-1)

    output_path = os.path.join(output_folder, f"frame_{frame_number}.jpg")
    
    cv2.imwrite(output_path, image)
    print(f"Saved: {output_path}")

print("Process complete!")
```
