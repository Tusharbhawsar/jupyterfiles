```python
def recuring_counts(input_list):
    counts = {}
    result = []
    
    for num in input_list:
        if num in counts:
            counts[num] += 1
        else:
            counts[num] = 1

    for count in counts.values():
        result.append(count)
    return result

input_list = [2,2,4,6,1,1,8,9,9,9]
result_list = recuring_counts(input_list)
print(result_list)
```
