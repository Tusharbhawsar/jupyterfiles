```python
import os
import pandas as pd
import urllib.request
import threading

# Read the CSV file
df = pd.read_csv("Baseball-Multi-Event-Classification-23-05-2024-16-06-52.csv")

# Extract URLs for each tag
ground = df[df['Tags'] == "ground"]['Source Url']
other = df[df["Tags"] == "other"]["Source Url"]
player = df[df["Tags"] == "player"]["Source Url"]
pitch = df[df["Tags"] == "pitch"]["Source Url"]
celebration = df[df["Tags"] == "celebration"]["Source Url"]
banner = df[df["Tags"] == "banner"]["Source Url"]

ground.shape
other.shape
player.shape
celebration.shape
banner.shape
pitch.shape

# Make directories if they don't exist
directories = ["ground", "other", "player", "pitch", "celebration", "banner"]
for directory in directories:
    if not os.path.exists(directory):
        os.mkdir(directory)

# Function to download images
def download_images(tag_urls, folder_name):
    for url in tag_urls:
        try:
            filename = os.path.join(folder_name, os.path.basename(urllib.parse.urlparse(url).path))
            urllib.request.urlretrieve(url, filename)
            print(f"Downloaded {filename}")
        except Exception as e:
            print(f"Failed to download {url}: {e}")

# Create threads for each category
threads = []
categories = [
    (ground, "ground"),
    (other, "other"),
    (player, "player"),
    (pitch, "pitch"),
    (celebration, "celebration"),
    (banner, "banner")
]

for urls, folder in categories:
    thread = threading.Thread(target=download_images, args=(urls, folder))
    threads.append(thread)
    thread.start()

# Wait for all threads to complete
for thread in threads:
    thread.join()
```
