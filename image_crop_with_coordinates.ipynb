```python
import glob
import cv2
import numpy as np
import urllib.request
import time
import json
import os

def download_url():

    #loading json data
    with open("helmat.json") as r:
        data_n = json.load(r)

    for i,item in enumerate(data_n):
        try:
            image=item['imageUrl']
            req = urllib.request.urlopen(image)
            arr = np.asarray(bytearray(req.read()), dtype=np.uint8)
            img = cv2.imdecode(arr, -1) # Load img as it is
            print(img.shape)

            # Cropping an image

            #cord = item['annotations']['rect'][0:4]
            

            if item['annotations']['rect'][4]=='false':

                cord = item['annotations']['rect'][0:4]
                x1 = cord[0]
                y1 = cord[1]
                x2 = cord[0]+ cord[2]
                y2 = cord[1]+ cord[3]

                cropped_image = img[y1:y2,x1:x2]

                h,w=cropped_image.shape[0],cropped_image.shape[1]
                #print(h,w)
                if (w==0)&(h==0):
                    continue

                print("cropped image size len",cropped_image.shape)

                output_path ="/home/link-lap-24/jupyter_files/dat/"
                cv2.imwrite(os.path.join(output_path, '{}.png'.format(i)),cropped_image)
               
            else:
                if item['annotations']['rect'][4]=='shooting':


                    cord = item['annotations']['rect'][0:4]



                    x1 = cord[0]
                    y1 = cord[1]
                    x2 = cord[0]+ cord[2]
                    y2 = cord[1]+ cord[3]

                    cropped_image = img[y1:y2,x1:x2]

                    h,w=cropped_image.shape[0],cropped_image.shape[1]
                    print(h,w)
                    if (w==0):
                        
                        continue

                    print("cropped image size len",cropped_image.shape)

                    output_path ="/home/link-lap-24/jupyter_files/dat/"
                    cv2.imwrite(os.path.join(output_path, '{}.png'.format(i)),cropped_image)


        except Exception as e:
            print(e)
download_url()
    
```
