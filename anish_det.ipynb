```python
import os
import shutil
import glob
import cv2
import requests
from termcolor import colored
import json
import os
# video_path = r""
# Frames = r"/home/multi-sy-007/Desktop/work/Basket_Ball/Frames"

DETECTION_URL = "http://localhost:5000/v1/object-detection/yolov5s"

# def files_in_folder(folder_path):
#     all_files = sorted(glob.glob(video_path + '/' + '*.png'))
#     #print(all_files)
#     for file in all_files:
#         shutil.move(file,Frames)

# files_in_folder(video_path)

#GET detection on images. Generate json response
def get_detections(IMAGE):
    with open(IMAGE, "rb") as f:
        image_data = f.read()
    response = requests.post(DETECTION_URL, files={"image": image_data}).json()
    return response

# Get results inj terms of co-ordinates
def get_cords_only(result):
    return [list(map(float,[i['xmin'], i['ymin'], i['xmax'], i['ymax'],(i['confidence'])])) for i in result]

# Draw Label & rectangle on image and save that image to new detection folder
def draw_box(image, result1, output_key,label):
    img_read = cv2.imread(image)
    Detection = Main_folder + "/df/"

    for res_i in result1:
        #print(type(res_i))

        x1, y1, x2, y2 = res_i['xmin'], res_i['ymin'], res_i['xmax'], res_i['ymax']
        #print(x1, y1, x2, y2)
        score=res_i['confidence']*100
        x1, y1, x2, y2 = list(map(int, [x1, y1, x2, y2]))
        # print(f'X1 {x1} : Y1 {y1} : X2 {x2} : Y2 {y2}')
        img_read = cv2.rectangle(img_read, (x1, y1), (x2, y2), (255, 0, 0), 2)
        img_read = cv2.putText(img_read, f"{score}", (x1, y1), cv2.FONT_HERSHEY_SIMPLEX, 2, 255)
    #print(Detection+output_key)
    tempT = cv2.imwrite(Detection+output_key, img_read)
    del tempT

def get_detection_on_frames(Main_folder):
    frame_folder = Main_folder + "/fr/"
    Detection = Main_folder + "/df/"

    all_detections = {}
    all_images = sorted(glob.glob(frame_folder + '*.png'), key=lambda x: int(x.split("/")[-1].split(".")[0]))
    #print(colored((all_images,blue))
    for index, image in enumerate(all_images):

        result1 = get_detections(image)
        print(result1)
        #print(colored(result1, "yellow"))
        #result1=[d for d in result1 if d.get('name') == "ball"]
        #result1 = [d for d in result1 if d.get('confidence') >=0.30]

        #print(colored(result1,"red"))

        '''
         0{'xmin': 566.3480224609, 'ymin': 562.5621337891, 'xmax': 594.3916015625, 'ymax': 587.9360351562,
          'confidence': 0.7584906816, 'class': 0, 'name': 'ball'}]
        '''
        if len(result1)==0:
            shutil.copy(image, Detection)
        else:
            out_key = os.path.basename(image)
            for result in result1:
                k=result["confidence"]
                label=result["class"]
                #print(label)
                #if label==0:
                    #if k >= 0.30 :
                    #result = sorted(result, key=operator.itemgetter("confidence"), reverse=True)
                    #result = [result[0]]

            draw_box(image, result1, out_key,label)
            all_detections[image.split("/")[-1]] = result1
             #print(all_detections)
    return all_detections

Main_folder= r"/home/link-lap-24/Downloads/yolo_testng"
frame_folder = Main_folder + "/fr/"
#print(frame_folder)
all_d = get_detection_on_frames(Main_folder)
json.dump(all_d, open(f"{Main_folder + '/Results/' + 'all_det_1.json'}", 'w'))
```
