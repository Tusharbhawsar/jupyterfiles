```python
import pandas as pd
import subprocess

# Read the CSV file
df = pd.read_csv('fr.csv')

# Iterate through each row in the CSV
for index, row in df.iterrows():
    print(row)
    
    tag = row['Type']
    source_url = row['Source Url']
    start_duration = row['startDuration']
    end_duration = row['endDuration']

    # Generate output filename
    output_filename = f"{tag}_{index}.mp4"

    # Run FFmpeg command
    cmd = [
        'ffmpeg',
        '-i', source_url,
        '-ss', str(start_duration),
        '-to', str(end_duration),
        '-c:v', 'libx264',
        '-c:a', 'aac',
        output_filename
    ]

    # Execute FFmpeg command
    subprocess.run(cmd)

print("Video trimming completed.")
```

```python
class Event:
    def __init__(self, type, date_scraped, date_of_event, time, link, blurb):
        countdown_delta = date_of_event - date_scraped
        countdown = countdown_delta.days

        if countdown < 0:
            has_passed = True
        else:
            has_passed = False

        self.type = type
        self.date_scraped = date_scraped
        self.date_of_event = date_of_event
        self.time = time
        self.link = link
        self.countdown = countdown
        self.has_passed = has_passed
        self.blurb = blurb

    def get_dictionary(self):
        event_dict = {}
        event_dict['type'] = self.type
        event_dict['scraped'] = self.date_scraped
        event_dict['date'] = self.date_of_event
        event_dict['time'] = self.time
        event_dict['url'] = self.link
        event_dict['countdown'] = self.countdown
        event_dict['blurb'] = self.blurb
        event_dict['has_passed'] = self.has_passed
        return event_dict
```

```python
import os
import cv2
import pandas as pd
import glob
import natsort
from ultralytics import YOLO

class AutoFlip:
    def __init__(self, video_path, model_path, output_dir):
        self.video_path = video_path
        self.model_path = model_path
        self.output_dir = output_dir
        self.frames_dir = os.path.join(output_dir, 'frames')
        self.audio_path = os.path.join(output_dir, 'audio.mp3')
        self.cropped_video_path = os.path.join(output_dir, 'cropped_video.mp4')
        self.final_output_path = os.path.join(output_dir, 'final_output.mp4')

    def frames_ext(self):
        if not os.path.exists(self.frames_dir):
            os.mkdir(self.frames_dir)
        cmd = f"ffmpeg -i {self.video_path} -q:v 1 {self.frames_dir}/%d.jpg"
        os.system(cmd)

    def extract_audio(self):
        cmd = f"ffmpeg -i {self.video_path} -q:a 0 -map a {self.audio_path}"
        os.system(cmd)

    def hei_weight(self):
        for index, filename in enumerate(os.listdir(self.frames_dir)):
            img = cv2.imread(os.path.join(self.frames_dir, filename))
            height, weight, _ = img.shape
            if index == 0:
                break
        return height, weight

    def det_frame(self):
        height, weight = self.hei_weight()

        start_point = int(9/16 * height)
        ext_st_time = int(start_point / 2)
        right_manu = start_point + start_point

        model = YOLO(self.model_path)
        bbox_data = []

        complete_path = natsort.natsorted(glob.glob(self.frames_dir + "/*.jpg"))
        for frame in complete_path:
            img_name = os.path.basename(frame)
            result = model(frame, max_det=1, conf=0.7)
            frame_bbox_data = {"frame_number": img_name, "top": None, "bottom": None, "left": None, "right": None}

            if result:
                for r in result:
                    boxes = r.boxes
                    for box in boxes:
                        b = box.xyxy[0].tolist()
                        x1, y1, x2, y2 = b[0], b[1], b[2], b[3]
                        row = int((y1 + y2) / 2)
                        column = int((x1 + x2) / 2)
                        top = 0
                        bottom = height
                        left = int(abs(column - ext_st_time))
                        right = int(left + start_point)

                        frame_bbox_data["top"], frame_bbox_data["bottom"], frame_bbox_data["left"], frame_bbox_data["right"] = top, bottom, left, right

                    bbox_data.append(frame_bbox_data)

        bbox_df = pd.DataFrame(bbox_data)
        bbox_df_interpolate = bbox_df.interpolate(method="linear", limit_direction="forward")
        bbox_df_interpolate = bbox_df_interpolate.fillna({"top": 0, "bottom": height, "left": start_point, "right": right_manu})
        bbox_df_interpolate_dict = bbox_df_interpolate.to_dict(orient="records")

        out = cv2.VideoWriter(self.cropped_video_path, cv2.VideoWriter_fourcc(*"mp4v"), 30, (start_point, height))
        for i, frame in enumerate(complete_path):
            img = cv2.imread(frame)
            top = int(bbox_df_interpolate_dict[i]["top"])
            bottom = int(bbox_df_interpolate_dict[i]["bottom"])
            left = int(bbox_df_interpolate_dict[i]["left"])
            right = int(bbox_df_interpolate_dict[i]["right"])
            cropping_frame = img[top:bottom, left:right]

            text = f"{i}"
            org = (50, 50)
            font = cv2.FONT_HERSHEY_SIMPLEX
            fontScale = 1
            color = (255, 0, 0)
            thickness = 2
            img_with_text = cv2.putText(cropping_frame, text, org, font, fontScale, color, thickness, cv2.LINE_AA)

            out.write(img_with_text)

        out.release()
        cv2.destroyAllWindows()

    def merge_audio_video(self):
        cmd = f"ffmpeg -i {self.cropped_video_path} -i {self.audio_path} -c copy -map 0:v:0 -map 1:a:0 {self.final_output_path}"
        os.system(cmd)

    def process_video(self):
        self.frames_ext()
        self.extract_audio()
        self.det_frame()
        self.merge_audio_video()

if __name__ == "__main__":
    video = "/home/link-lap-24/Autoflip_verticle/auto_flip_test2/2.mp4"
    model_path = "/home/link-lap-24/Autoflip_verticle/auto_flip_test2/5uY7XAMfU35Gylj_golfball_v8.pt"
    output_dir = "/home/link-lap-24/Autoflip_verticle/auto_flip_test2/"

    autoflip = AutoFlip(video, model_path, output_dir)
    autoflip.process_video()
```
