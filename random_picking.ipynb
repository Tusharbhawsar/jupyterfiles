```python
import uuid
import urllib
import pandas as pd
import random
import glob
import natsort
import os
```

```python
data=pd.read_csv("3_pointer.csv")

df = data.sample(frac = 1)

df
```

```python
for i in df['Source Url']:
#     print(i)
    name=str(uuid.uuid4())
    
    urllib.request.urlretrieve(i,name)
```

```python
video_path="/home/link-lap-24/Downloads/3pointer/"

cmp_path=natsort.natsorted(glob.glob(video_path + "*.mp4"))
cmp_path
```

```python
for i in cmp_path:
    vid_name=i.split(".")[0]
    cr_path=os.path.join(video_path,vid_name)
    os.makedirs(cr_path)
    cmd=f"""ffmpeg -i {i} -qscale:v 1 {cr_path}/%d.png"""
    os.system(cmd)
```
