```python
import natsort, os, shutil
```

# Data Preprocessing

```python
primary_set_path = "DATASET-1/"
secondary_set_path = "RealBBOX/"
```

```python
# print(primary_set_path, secondary_set_path)
```

```python
primarySet = natsort.natsorted(os.listdir(primary_set_path))
```

```python
secondarySet = natsort.natsorted(os.listdir(secondary_set_path))
```

```python
print("PrimarySet_images::::",len(primarySet), "SecondarySet_images::::", len(secondarySet))
```

```python
primarySetIds = []

p_count = 0
for img in primarySet:
    string = img.split("-")
    string = string[:len(string)-2]
    if len(string) == 1:
        vid = string[0]
    else:
        vid = "-".join(string)
    if vid not in primarySetIds:
        primarySetIds.append(vid)
        p_count += 1
        print(img)
        print("vid::", vid, "inx::", p_count)
        print()
    if vid in primarySetIds:
        continue
```

```python
secondarySetIds = []

s_count = 0
for img in secondarySet:
    string = img.split("-")
    string = string[:len(string)-2]
    if len(string) == 1:
        vid = string[0]
    else:
        vid = "-".join(string)

    if vid not in secondarySetIds:
        secondarySetIds.append(vid)
        s_count += 1
        print(img)
        print("vid::", vid, "inx::", s_count)
        print()
    
    if vid in secondarySetIds:
        continue
```

# DATA GATHERING

```python
input_video = "https://d1lw40p9dgniy0.cloudfront.net/vendor/pc9tHHBZf/1649509311.mp4"
video_id = input_video.split("/")[-1].split(".")[0]
print(video_id)

outputDir = video_id + "/"
if not os.path.exists(outputDir):
    os.mkdir(outputDir)
    
    
dataset = ""
source = ""

if video_id in primarySetIds:
    dataset = primarySet
    source = primary_set_path
    
if video_id in secondarySetIds:
    dataset = secondarySet
    source = secondary_set_path
    
print("source", source)


def collect_data(dataset, video_id, source):
    for img in dataset:
        if video_id in img:
            print(img)
            ip_path = source + img
            op_path = outputDir + img
            print(ip_path, op_path)
            
            shutil.copy(ip_path, op_path)            
collect_data(dataset, video_id, source)
```
