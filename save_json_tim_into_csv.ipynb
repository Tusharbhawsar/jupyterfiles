```python
def decimal_to_time_with_milliseconds(decimal_number):
    # Extract hours, minutes, seconds, and milliseconds
    hours = int(decimal_number / 60 / 60)
    minutes = int((decimal_number / 60) % 60)
    seconds = int(decimal_number % 60)
    milliseconds = int((decimal_number - int(decimal_number)) * 1000)

    # Format components with two digits
    formatted_hours = str(hours).zfill(2)
    formatted_minutes = str(minutes).zfill(2)
    formatted_seconds = str(seconds).zfill(2)
    formatted_milliseconds = str(milliseconds).zfill(3)

    # Combine components into the "00:00:00.000" format
    result = f"{formatted_hours}:{formatted_minutes}:{formatted_seconds}.{formatted_milliseconds}"
    return result

# Example usage
decimal_number = 295.002
formatted_time = decimal_to_time_with_milliseconds(decimal_number)
print(formatted_time)
```

```python
## prabhansh

import json
import datetime
import csv

def format_duration(seconds):
    whole_seconds = int(seconds)
    milliseconds = int((seconds - whole_seconds) * 1000)
    return f"{whole_seconds}.{milliseconds:03d}"

file_path = '/home/link-lap-24/exp/Volleyball-Gt-Replay-24-01-2024-17-43-22.json'
with open(file_path, 'r') as file:
    json_data = json.load(file)

formatted_data = []

for entry in json_data:
    source_url = entry["sourceUrl"]

    for event in entry["event"]:
        formatted_start_time = format_duration(event["startTime"])
        formatted_end_time = format_duration(event["endTime"])
        formatted_time_range = f"{formatted_start_time}-{formatted_end_time}"
        formatted_data.append({
            "SourceUrl": source_url,
            "TimeRange": formatted_time_range,
            "Event": event["event"]
        })

csv_file_path = "formatted_data.csv"
fields = ["SourceUrl", "TimeRange", "Event"]

with open(csv_file_path, mode='w', newline='') as csv_file:
    writer = csv.DictWriter(csv_file, fieldnames=fields)
    writer.writeheader()

    writer.writerows(formatted_data)

print(f"Formatted data saved to {csv_file_path}")
```
