```python
import os
import uuid

file1 = open('negative.txt', 'r')
count = 0

path = "/home/link-lap-24/Downloads/manish_sir/"

while True:

    count += 1

    random = str(uuid.uuid4())

    line = file1.readline()

    if not line:
        break

    print(path, random)
    #os.mkdir(random)

    cmd  = "ffmpeg -i {} -vf fps=1 {}{}_%d.jpg".format(line.strip(),path,random)
    
    print(cmd)
    os.system(cmd)
 
file1.close()
```
